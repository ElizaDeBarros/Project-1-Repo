```python
import requests
import json
import pandas as pd
```

```python
# pull specific url for texas wind farm data
url = "https://eersc.usgs.gov/api/uswtdb/v1/turbines?&t_state=eq.TX"
```

```python
# ensure request goes through
print(requests.get(url))
```

```python
# create response variable in json format
response = requests.get(url).json()
```

```python
# print new dataframe for texas wind farm data
texas_df = pd.DataFrame(response)
texas_df
```

```python
# drop all unnecessary columns
texas_df = texas_df.drop(columns=['t_state','case_id','faa_ors','faa_asn','usgs_pr_id','t_fips','t_rsa','t_cap','t_ttlh','t_conf_atr', 't_conf_loc','t_img_date', 't_img_srce','eia_id', 'retrofit','retrofit_year'])

texas_df
```

```python
# rename columns for cleaner look
texas_df = texas_df.rename(columns={'t_county':'County',
                               'p_name':'Wind Farm Name',
                               'p_year': 'Year',
                               'p_tnum':'Turbine Count',
                               'p_cap':'Cumulative Capacity of Wind Farm (in MW)',
                               't_manu':'Turbine Manufacturer',
                               't_model':'Turbine Model',
                               't_hh': 'Turbine Hub Height',
                               't_rd':'Rotor Diameter',
                               'xlong':'Longitude',
                               'ylat':'Latitude'
                              })
texas_df
```

```python
# drop all years that have null values
texas_df.dropna(subset=["Year"], inplace=True)
texas_df
```

```python
# drop all years that are older than 2000
current_tx_df = texas_df[texas_df["Year"] >= 2000]
current_tx_df
```

```python
# running some simple stats on energy output
current_tx_df["Cumulative Capacity of Wind Farm (in MW)"].describe()
```

```python
# showing all counties in texas that have wind farms 
current_tx_df["County"].value_counts()
```

```python
# print clean texas csv file for graphs and statistical analyses
current_tx_df.to_csv("tx_clean.csv",
                  encoding="utf-8", index=False, header=True)
```
