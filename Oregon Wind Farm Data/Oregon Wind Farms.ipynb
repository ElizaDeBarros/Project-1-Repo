```python
import requests
import json
import pandas as pd
```

```python
url = "https://eersc.usgs.gov/api/uswtdb/v1/turbines?&t_state=eq.OR"
```

```python
print(requests.get(url))
```

```python
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
oregon_df = pd.DataFrame(response)
oregon_df
```

```python
per_year = oregon_df["p_year"].value_counts().sort_values(ascending=False)
per_year
```

```python
oregon_df = oregon_df.drop(columns=['case_id','faa_ors','faa_asn','usgs_pr_id','t_fips','t_rsa','t_cap','t_ttlh','t_conf_atr', 't_conf_loc','t_img_date', 't_img_srce','eia_id', 'retrofit','retrofit_year'])

oregon_df
```

```python
clean_or_df = oregon_df[["t_county", "p_name", "p_year", "p_tnum", "p_cap", "xlong", "ylat"]]
clean_or_df
```

```python
or_df = clean_or_df.rename(columns={"t_county": "County", 
                                    "p_name": "Wind Farm Name", 
                                    "p_year": "Year", 
                                    "p_tnum": "Turbine Count", 
                                    "p_cap": "Cumulative Capacity of Wind Farm (in MW)", 
                                    "xlong": "Longitude", 
                                    "ylat": "Latitude"
                                   })
or_df
```

```python
or_df.dropna(subset=["Year"], inplace=True)
or_df
```

```python
current_or_df = or_df[or_df["Year"] >= 2000]
current_or_df
```

```python
current_or_df["Cumulative Capacity of Wind Farm (in MW)"].describe()
```

```python
current_or_df["Cumulative Capacity of Wind Farm (in MW)"].value_counts()
```

```python
current_or_df.to_csv("or_clean.csv",
                  encoding="utf-8", index=False, header=True)
```
