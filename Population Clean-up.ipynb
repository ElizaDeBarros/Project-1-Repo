```python
# Dataframe list: PopTX, PopOR, PopOK, PopIA, Employment, PopUS
# Pop dataframe clean up tasks: Add state column, change column headings
# Employment clean up tasks: delete a bunch of columns, rename columns, 
    # seperate out state from county name, filter to OR, OK, IA, TX
# PopUS clean up: delete a bunch of columns, rename columns, filter to OR, OK, IA, TX
```

```python
def pop_clean(popstate, state):
    df = popstate.rename(columns={"Geographic Area": "County",
                                  "2000": state+ " Pop 2000",
                                  "2001": state+ " Pop 2001",
                                  "2002": state+ " Pop 2002",
                                  "2003": state+ " Pop 2003",
                                  "2004": state+ " Pop 2004",
                                  "2005": state+ " Pop 2005",
                                  "2006": state+ " Pop 2006",
                                  "2007": state+ " Pop 2007",
                                  "2008": state+ " Pop 2008",
                                  "2009": state+ " Pop 2009"})
    df["State"] = state
    
    df = df[["State", "County", state+ " Pop 2000",
                                  state+ " Pop 2001",
                                  state+ " Pop 2002",
                                  state+ " Pop 2003",
                                  state+ " Pop 2004",
                                  state+ " Pop 2005",
                                  state+ " Pop 2006",
                                  state+ " Pop 2007",
                                  state+ " Pop 2008",
                                  state+ " Pop 2009"]]
    df["County"] = df["County"].replace({".":""})
    
        
    return df
```

```python
PopOK_clean = pop_clean(PopOK,'OK')
PopOK_clean.to_csv("Data/Data_Output/PopOK.csv", encoding="utf-8", index=False, header=True)
PopOK_clean.head()
```

```python
PopTX_clean = pop_clean(PopTX,'TX')
PopTX_clean.to_csv("Data/Data_Output/PopTX.csv", encoding="utf-8", index=False, header=True)
PopTX_clean.head()
```

```python
PopOR_clean = pop_clean(PopOR,'OR')
PopOR_clean.to_csv("Data/Data_Output/PopOR2.csv", encoding="utf-8", index=False, header=True)
PopOR_clean.head()
```

```python
PopIA_clean = pop_clean(PopIA,'IA')
PopIA_clean.to_csv("Data/Data_Output/PopIA.csv", encoding="utf-8", index=False, header=True)
PopIA_clean.count()
```

```python
PopUS_renamed = PopUS.rename(columns={"STNAME": "State",
                                  "CTYNAME": "County",
                                  "CENSUS2010POP": "Pop 2010",
                                  "POPESTIMATE2011": "Pop 2011",
                                  "POPESTIMATE2012": "Pop 2012",
                                  "POPESTIMATE2013": "Pop 2013",
                                  "POPESTIMATE2014": "Pop 2014",
                                  "POPESTIMATE2015": "Pop 2015",
                                  "POPESTIMATE2016": "Pop 2016",
                                  "POPESTIMATE2017": "Pop 2017",
                                  "POPESTIMATE2018": "Pop 2018",
                                  "POPESTIMATE2019": "Pop 2019",})

PopUS_Clean = PopUS_renamed[["State", "County", "Pop 2010", "Pop 2011", "Pop 2012",
                                  "Pop 2013", "Pop 2014", "Pop 2015", "Pop 2016",
                                  "Pop 2017", "Pop 2018", "Pop 2019"]]

PopUS_Clean.to_csv("Data/Data_Output/PopUS.csv", encoding="utf-8", index=False, header=True)
PopUS_Clean.head()
```

```python
# Cut PopUS to 4 states of interest

states = ['Oregon', 'Iowa', 'Texas', 'Oklahoma']

for state in states:
    PopUS_4states = PopUS_Clean.loc[PopUS_Clean['State'] == state]

PopUS_4states['State'] = PopUS_4states['State'].replace({"Oklahoma": "OK",
                                                        "Iowa": "IA",
                                                        "Oregon": "OR",
                                                        "Texas": "TX"})
PopUS_4states.to_csv("Data/Data_Output/Pop4states.csv", encoding="utf-8", index=False, header=True)
PopUS_4states.head()
```
