```python
import requests
import json
import pandas as pd
```

```python
url = "https://eersc.usgs.gov/api/uswtdb/v1/turbines?&t_state=in.(CA)&select=t_county,p_name,p_year,p_tnum,p_cap,xlong,ylat"
```

```python
print(requests.get(url))
```

```python
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
cali_df = pd.DataFrame(response)
cali_df
```

```python
clean_cali_df = cali_df.rename(columns={"t_county": "County", 
                                    "p_name": "Wind Farm Name", 
                                    "p_year": "Year", 
                                    "p_tnum": "Turbine Count", 
                                    "p_cap": "Cumulative Capacity of Wind Farm (in MW)", 
                                    "xlong": "Longitude", 
                                    "ylat": "Latitude"
                                   })
clean_cali_df
```

```python
clean_cali_df.dropna(subset=["Year"], inplace=True)
clean_cali_df
```

```python
current_cali_df = clean_cali_df[clean_cali_df["Year"] >= 2000]
current_cali_df
```

```python
current_cali_df["Cumulative Capacity of Wind Farm (in MW)"].describe()
```

```python
current_cali_df["Cumulative Capacity of Wind Farm (in MW)"].value_counts()
```

```python
current_cali_df.to_csv("cali_clean.csv",
                  encoding="utf-8", index=False, header=True)
```
